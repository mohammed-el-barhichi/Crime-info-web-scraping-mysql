# Importing packages

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

# Extracting data

```python
html_text = requests.get('https://jamescitycountyva.gov/CivicAlerts.aspx?CID=13').text
soup = BeautifulSoup(html_text, 'lxml')
wanteds = soup.find_all('div', class_='item intro fr-view')
for wanted in wanteds[2:]:
    name = wanted.find('h3').text
    posted_date = wanted.find('span', {'class': 'date'}).text
    detail = wanted.find_all('p')[1]
    detail_text = detail.text
    page = wanted.find('a')['href']
    image = wanted.find('img')['src']
    print(f'''
    Wanted name: {name}
    {posted_date}
    Details: {detail_text}
    Link: https://jamescitycountyva.gov/{page}
    Image: https://jamescitycountyva.gov/{image}
    ''')
```

```python
df = pd.DataFrame(columns=["Wanted name", "Date", "Image", "Link to page", "Details"])

html_text = requests.get('https://jamescitycountyva.gov/CivicAlerts.aspx?CID=13').text
soup = BeautifulSoup(html_text, 'lxml')
wanteds = soup.find_all('div', class_='item intro fr-view')
for wanted in wanteds[2:]:
    name = wanted.find('h3').text
    posted_date = wanted.find('span', {'class': 'date'}).text[11:]

    detail = wanted.find_all('p')[1].text
    page = "https://jamescitycountyva.gov"+wanted.find('a')['href']
    img = "https://jamescitycountyva.gov"+wanted.find('img')['src']
    df = df._append({"Wanted name":name, "Date":posted_date, "Image":img, "Link to page":page, "Details":detail}, ignore_index=True)


df['Date'] = pd.to_datetime(df['Date'])

df.head()

```
